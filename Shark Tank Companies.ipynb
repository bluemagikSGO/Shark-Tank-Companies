### PLOTTING A BAR CHART TO KNOW WHICH ENTREPRENUE GETS THE HIGHEST CAPITAL AND ONE WITH THE LOWEST CAPITAL

```python
x = accepted_s1["entrepreneurs"]
y = accepted_s1["askedfor"]
```

```python
plt.bar(x,y, color = 'b')
bars = plt.bar(x, y)
plt.xticks(rotation = 'vertical')
plt.title("HIGHEST/LOWEST CAPITAL DISBURSE (season 01)")
plt.xlabel("entreprenues")
plt.ylabel("capital invested($)")
bars[2].set_hatch('*')
bars[21].set_hatch('*')
bars[18].set_hatch('+')
plt.show()
```

# CHART ANALYSIS: from the above BAR CHART, we can deduce that in season one, [susan knapp] & [Marix Stone and Dr. Nancy Tanchel] got the highest capital from the sharks worth 500,000 (dollars) each, while the exchange is for a 15% and 20% stake respectively.
# The entreprenue that got the lowest capital is [Dr. Floyd Seskin] worth 25,000 (dollars) with the exchange for stake being 51%.
